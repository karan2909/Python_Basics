```python
def sqInRect(x, y):
    if x == y:
        return None
        
    result = []
    while y != x:
        if y > x:
            y = y- x
            result.append(x)
        elif x > y:
            x = x- y
            result.append(y)

    result.append(y)
    
    return result
```

```python
sqInRect(3,5)
```
